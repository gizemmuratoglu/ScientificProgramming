```python
from collections import Counter
data = ""
count = 0
f = open('shakespeare.txt', 'r')
for i in f:
    data += i
txt = Counter(data.split())
for i in txt:
    if txt.get(i) > 4:
        count += 1
print(count)
```

```python
from collections import Counter
data = ""
file = open('shakespeare.txt', 'r')
for i in file:
    data += i
data = data.split()
txt = str(Counter(data).most_common(200))
f = open('mostcommon200.txt', 'w')
f.write(txt)
f.close()
```

```python
import random
def data_generation(n, a, b):
    mylist = []
    max_val=max(a, b)
    min_val =  min(a, b)
    for i in range(n):
        mylist.append(random.randint(min_val, max_val))
    f = open('generation.txt', 'w')
    f.write(str(mylist))
    f.close()

data_generation(5, 14, 1)
```
